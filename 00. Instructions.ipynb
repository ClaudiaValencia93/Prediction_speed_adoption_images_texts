###Instrucciones:

En este proyecto se tienen seis notebooks, los cuales son:

1. Descriptive-analysis_and_first_iteration_classical_machine_learning
2. Second_iteration_classical_machine_learning
3. Multiclass_image_classification_deep_learning
4. Multiclass_text_classification_deep_learning
5. Principal_components_analysis_PCA
6. Sequential_Feature_Selector_and_classical_machine_learning

Los notebooks se pueden ejecutar linealmente sin ninguna complicación, sin embargo cuando se distribuyen las imágenes por clases en diferentes carpetas y cuando se realiza el preprocesamiento de las imágenes es pesado y tarda horas en ejecutarse.

El dataset se descarga desde una ruta compartida en drive o lo pueden encontrar en la página de Kaggle: https://www.kaggle.com/c/petfinder-adoption-prediction/data.

Cada uno de los pasos en los seis notebooks están documentados y son fáciles de comprender. Además, tienen una explicación amplia en el trabajo escrito.
